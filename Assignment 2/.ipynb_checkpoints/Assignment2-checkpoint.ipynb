*__a. Find a linear regression model that relates the living area to the selling price.
If you did any data cleaning step(s), describe what you did and explain why.__*

***

Let's start off by importing the data for the assignment.

```python
import pandas as pd

df_villas = pd.read_csv('data/data_assignment2.csv')

print(df_villas)
```

We need to build a linear regression model that relates the living area to the sell price. Let's see if any of those values are missing.

```python
df_villas[['Living_area', 'Selling_price']].isnull().any()
```

Neither of the columns contain missing values, so no data cleaning is required.

Let's visualize the data.

```python
import matplotlib.pyplot as plt

plt.scatter(
    df_villas['Living_area'],
    df_villas['Selling_price'],
)

plt.show()
```

At a first glance, the living area seems correlated to the selling price of the property. There seem to only be a couple of more significant outliers.

Let's build our model and draw the regression line.

```python
import numpy as np
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(df_villas['Living_area'].array[:, np.newaxis], df_villas['Selling_price'])

living_area_fit = np.array([60, 250])
selling_price_fit = model.predict(living_area_fit[:, np.newaxis])

plt.scatter(
    df_villas['Living_area'],
    df_villas['Selling_price'],
)
plt.plot(living_area_fit, selling_price_fit)

plt.show()
```

***

*__b. What are the values of the slope and intercept of the regression line?__*

***

```python
print(f'Value of the slope: {model.coef_[0]}')
print(f'Value of the intercept: {model.intercept_}')
```

***

*__c. Use this model to predict the selling prices of houses which have living area 100 m2, 150 m2 and 200 m2.__*

***

```python
living_areas = np.array([100, 150, 200])
predicted_prices = model.predict(living_areas[:, np.newaxis])

for living_area, predicted_price in zip(living_areas, predicted_prices):
    print(f'Predicted price for villas with a living area of {living_area} m2: {predicted_price}')
```

***

The points in the residual plot show no discernible pattern around the residual = 0 line, which indicates that a linear model is a good fit for the data.

***

*__In this question, you will use the Iris data set (“from sklearn.datasets import
load_iris”).__*

*__a. Use a confusion matrix to evaluate the use of logistic regression to classify the
iris data set.__*

***

```python
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split

# Import iris data
iris = load_iris(as_frame=True)
x = iris.data[["sepal length (cm)", "sepal width (cm)"]]
y = iris.target
class_names = iris.target_names

# Split the data into a training set and a test set
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)

# Run classifier
classifier = LogisticRegression(max_iter=1000)
classifier.fit(x_train, y_train)

# Plot non-normalized and normalized confusion matrix
plot_options = [
    ("Confusion matrix, without normalization", None),
    ("Normalized confusion matrix", "true"),
]

for title, normalize in plot_options:
    disp = ConfusionMatrixDisplay.from_estimator(
        classifier,
        x_test,
        y_test,
        display_labels=class_names,
        cmap=plt.cm.Blues,
        normalize=normalize,
    )
    disp.ax_.set_title(title)

plt.show()
```

Based on the confusion matrix, logistic regression is a decent fit for classifying the iris dataset.

***

*__b. Use k-nearest neighbours to classify the iris data set with some different
values for k, and with uniform and distance-based weights. What will happen
when k grows larger for the different cases? Why?__*

***

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.inspection import DecisionBoundaryDisplay

# Define values for K and types of weights
k_list = [1, 5, 10, 15, 20]
weights_list = ['uniform', 'distance']

combinations = np.array(np.meshgrid(k_list, weights_list)).T.reshape(-1, 2)

# Run classifier for every possible combination
for k, weights in combinations:
    classifier = KNeighborsClassifier(n_neighbors=int(k))
    classifier.set_params(weights=weights).fit(x_train, y_train)
    disp = DecisionBoundaryDisplay.from_estimator(
        classifier,
        x_test,
        response_method="predict",
        plot_method="pcolormesh",
        xlabel=iris.feature_names[0],
        ylabel=iris.feature_names[1],
        shading="auto",
        alpha=0.5
    )
    scatter = disp.ax_.scatter(x.iloc[:, 0], x.iloc[:, 1], c=y, edgecolors="k")
    disp.ax_.legend(
        scatter.legend_elements()[0],
        iris.target_names,
        loc="lower left",
        title="Classes",
    )
    disp.ax_.set_title(
        f"3-Class classification\n(k={k}, weights={weights!r})"
    )

plt.show()
```

We can see that by increasing the value of K, the number of clusters decreases and the clusters are better defined. The reason is that by increasing the value of K, outliers are less likely to impact the classifier. When using "distance" weights, some additional clusters are observed near isolated outliers.

***

*__c. Compare the classification models for the iris data set that are generated by
k-nearest neighbours (for the different settings from question b) and by
logistic regression. Calculate confusion matrices for these models and discuss
the performance of the various models__*

***

```python
# Calculate confusion matrix for every possible combination
for k, weights in combinations:
    classifier = KNeighborsClassifier(n_neighbors=int(k))
    classifier.set_params(weights=weights).fit(x_train, y_train)
    disp = ConfusionMatrixDisplay.from_estimator(
        classifier,
        x_test,
        y_test,
        display_labels=class_names,
        cmap=plt.cm.Blues
    )
    title = f'Confusion matrix\n(k={k}, weights={weights!r})'
    disp.ax_.set_title(title)

plt.show()
```

When comparing the confusion matrices for every configuration of KNN with the confusion matrix for the logistic regression that we calculated earlier, we can see that we are getting worse results when using KNN for most configurations. The best results are achieved with K = 20 and weights = 'uniform', which outperform the logistic regression as well.
